```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Salary'], kde=True, bins=30, color='skyblue')
plt.title('Probability Density Function (PDF) and Histogram for Salary')
plt.xlabel('Salary')
plt.ylabel('Density')
plt.show()
```

```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1
outliers_salary = df[(df['Salary'] < Q1 - 1.5 * IQR) | (df['Salary'] > Q3 + 1.5 * IQR)]
print("Outliers in Salary:")
print(outliers_salary)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Salary'], color='lightcoral')
plt.title('Boxplot for Salary')
plt.xlabel('Salary')
plt.show()
```
